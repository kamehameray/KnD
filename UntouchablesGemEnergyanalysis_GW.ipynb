```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid")

#import scatterplot data csv
gem = pd.read_csv('./data/Untouchables_Energy_Gemscatter.csv',delimiter = ',')
gemavg = pd.read_csv('./data/Untouchables_Energy_GemAVG.csv',delimiter = ',')
```

```python
fgrid = sns.stripplot(x ='gem_spent', y='energy_spent', data= gem, hue = 'min_yr')
fgrid.set_yscale('log')
```

```python
fgrid = sns.lmplot(x ='gem_spent', y='energy_spent', data= gem, hue = 'min_yr', fit_reg = False)
fgrid.set(xscale="log", yscale="log")
```

```python
fgrid = sns.lmplot(x ='gem_spent', y='energy_spent', data= gem, hue = 'max_yr', fit_reg = False)
fgrid.set(xscale="log", yscale="log")
```

```python
fgrid = sns.lmplot(x ='avg_gem_spent', y='avg_energy_spent', data= gemavg, hue = 'min_yr', fit_reg = False)
fgrid.set(xscale="log", yscale="log")
```

```python
fgrid = sns.lmplot(x ='avg_gem_spent', y='avg_energy_spent', data= gemavg, hue = 'max_yr', fit_reg = False)
fgrid.set(xscale="log", yscale="log")
```
